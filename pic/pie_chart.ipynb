```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib import style

# Apply FiveThirtyEight style
style.use('fivethirtyeight')

# Data for inner pie (sub-race distribution)
labels = ['White',
    'Hispanic or Latino', 'Black or African American',  'Asian',
    'Two or More Races', 'Native Hawaiian or Other Pacific Islanders', 'American Indian or Alaska Native'
]
sizes = [22.4, 41.1, 23.4, 5.55, 2.36, 0.232, 0.221]

# Merge categories <= 1% into 'Others'
threshold = 1.0
others_sum = sum(size for size in sizes if size <= threshold)
new_sizes = [size for size in sizes if size > threshold] + [others_sum]
new_labels = [label for label, size in zip(labels, sizes) if size > threshold] + ['Native Hawaiian or Other Pacific Islanders,\n and American Indian or Alaska Native']

# Aggregate for outer pie (Traditional vs. Non-Traditional)
traditional = sizes[0]  # White
non_traditional = sum(sizes[1:])  # All others

# Data for outer pie
outer_labels = ['Traditional', 'Non-Traditional']
outer_sizes = [traditional, non_traditional]

# Custom color palette: User-preferred colors
inner_colors = ['#003f5c', '#444e86', '#955196', '#dd5182', '#ff6e54', '#ffa600']  # Original sub-groups colors
outer_colors = ['#444e86', '#ff6e54']  # Similar shades for aggregated groups

# Plot
fig, ax = plt.subplots(figsize=(8, 6))

# Outer pie (aggregated categories)
wedges_outer, texts_outer, autotexts_outer = ax.pie(
    outer_sizes, colors=outer_colors, radius=1, 
    autopct='%1.1f%%', startangle=140, pctdistance=0.85, 
    wedgeprops=dict(width=0.3, edgecolor='w'))

# Inner pie (sub-race distribution)
wedges_inner, texts_inner, autotexts_inner = ax.pie(
    new_sizes, colors=[inner_colors[1]] + inner_colors[2:], radius=0.7, 
    autopct='%1.1f%%', startangle=140, pctdistance=0.75, 
    wedgeprops=dict(width=0.4, edgecolor='w'))

# Title
plt.title('Demographic Distribution \n(Mercy University)', fontsize=18, fontweight='bold', pad=20)

# Ensure aspect ratio is equal to make the pie chart circular
ax.set_aspect('equal')

# Improve text appearance on pie
plt.setp(autotexts_outer + autotexts_inner, size=12, weight="bold", color="white")

# Draw the plot
plt.tight_layout()

# Add legend at the bottom
plt.legend(wedges_inner + wedges_outer, new_labels + outer_labels, title="Demographics", 
           loc="upper center", bbox_to_anchor=(0.5, -0.01), 
           fontsize=12, title_fontsize=14, ncol=2)

plt.show()
```

## Merge

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib import style

# Apply FiveThirtyEight style
style.use('fivethirtyeight')

# Data for inner pie (sub-race distribution)
labels = ['White',
          'Hispanic or Latino', 'Black or African American',  'Asian',
          'Two or More Races', 'Native Hawaiian or Other Pacific Islanders', 'American Indian or Alaska Native'
          ]
sizes = [22.4, 41.1, 23.4, 5.55, 2.36, 0.232, 0.221]

# Merge categories <= 3% into 'Others'
threshold = 3.0
others_sum = sum(size for size in sizes if size <= threshold)
new_sizes = [size for size in sizes if size > threshold] + [others_sum]
new_labels = [label for label, size in zip(
    labels, sizes) if size > threshold] + ['Others']

# Aggregate for outer pie (Traditional vs. Non-Traditional)
traditional = sizes[0]  # White
non_traditional = sum(sizes[1:])  # All others

# Data for outer pie
outer_labels = ['Traditional', 'Non-Traditional']
outer_sizes = [traditional, non_traditional]

# Custom color palette: Darker shades for traditional and non-traditional
inner_colors = ['#ADD8E6', '#ffa600', '#ff6e54', '#dd5182',
                '#955196', '#444e86', '#003f5c', ]  # Original sub-groups colors
outer_colors = ['#301934', '#2c3e50']  # Darker shades for Traditional and Non-Traditional

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 8))

# Function to adjust the percentage labels
def custom_autopct(pct):
    return ('%1.1f%%' % pct) if pct > 0 else ''

# Plot for Race Distribution (left)
# Outer pie (aggregated categories)
wedges_outer, texts_outer, autotexts_outer = ax1.pie(
    outer_sizes, colors=outer_colors, radius=1,
    autopct=custom_autopct, startangle=140, pctdistance=0.85,
    wedgeprops=dict(width=0.3, edgecolor='w'))

# Inner pie (sub-race distribution)
wedges_inner, texts_inner, autotexts_inner = ax1.pie(
    new_sizes, colors=[inner_colors[1]] + inner_colors[2:], radius=0.7,
    autopct='%1.1f%%', startangle=140, pctdistance=0.75,
    wedgeprops=dict(width=0.4, edgecolor='w'))

# Ensure aspect ratio is equal to make the pie chart circular
ax1.set_aspect('equal')

# Improve text appearance on pie
plt.setp(autotexts_inner, size=12, weight="bold", color="white")
plt.setp(autotexts_outer, size=16, weight="bold", color="white")

# Add legend at the bottom of the left pie chart
ax1.legend(wedges_inner, new_labels, title="Demographics",
           loc="upper center", bbox_to_anchor=(0.5, -0.01),
           fontsize=16, title_fontsize=20, ncol=2)

# Data for Age Group Distribution (right)
age_groups = ['18-19', '20-21', '22-24', '25-29', '30-34', '35 and over']
amounts = [1409, 1436, 1642, 1702, 874, 1549]

# Aggregate the data into the new groups
traditional_group = sum([amounts[0], amounts[1], amounts[2]])
non_traditional_group = sum([amounts[3], amounts[4], amounts[5]])

# New data for the outer pie chart
group_labels = ['Traditional (18-24)', 'Non-Traditional (25+)']
group_amounts = [traditional_group, non_traditional_group]

# Plot for Age Group Breakdown (right)
# Inner ring for original sub-groups
wedges_inner, texts_inner, autotexts_inner = ax2.pie(amounts, radius=0.7, labels=None,
                                                     colors=inner_colors, autopct='%1.1f%%', pctdistance=0.85,
                                                     startangle=140, textprops={'fontsize': 12},
                                                     wedgeprops=dict(width=0.4, edgecolor='white'))

# Outer ring for aggregated groups
wedges_outer, texts_outer, autotexts_outer = ax2.pie(group_amounts, radius=1, labels=None,
                                                     colors=outer_colors, autopct=custom_autopct, pctdistance=0.85,
                                                     startangle=140, textprops={'fontsize': 14, 'weight': 'bold'},
                                                     wedgeprops=dict(width=0.3, edgecolor='white'))

# Improve text appearance
plt.setp(autotexts_inner, size=12, weight="bold", color="white")
plt.setp(autotexts_outer, size=16, weight="bold", color="white")

# Ensure aspect ratio is equal to make the pie chart circular
ax2.set_aspect('equal')

# Add legend at the bottom of the right pie chart
ax2.legend(wedges_inner, age_groups, title="Age",
           loc="upper center", bbox_to_anchor=(0.5, -0.01),
           fontsize=16, title_fontsize=20, ncol=2)

# Add third legend for Traditional vs Non-Traditional below both charts
fig.legend(wedges_outer, group_labels, title="Traditional vs Non-Traditional",
           loc="lower center", bbox_to_anchor=(0.5, -0.08),
           fontsize=16, title_fontsize=20, ncol=2)

# Adjust layout to accommodate the third legend
plt.tight_layout(rect=[0, 0.1, 1, 1])
plt.subplots_adjust(bottom=0.2)

# Display the combined figure
plt.show()
```
